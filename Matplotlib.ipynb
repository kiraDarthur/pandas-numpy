Matplotlib é uma biblioteca em Python utilizada para criação de gráficos e visualizações de dados. Ela oferece uma ampla gama de funcionalidades para criar diferentes tipos de gráficos, incluindo gráficos de linha, dispersão, barras, histogramas, pizza, entre outros.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y =[10,20,35,40,50,99]

fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python

import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# criando os dados 
x = np.linspace(0, 10, 1000)
y = 4 + 2 * np.sin(2 * x)

#  criando uma figura
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=3.0)# aqui é realmente  criando grafico

#aqui criando a figura
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 9))

plt.show()#exibindo a figura
```

```python
import pandas as pd
dados = {
    'x': [1, 2, 3, 4, 5],
'y': [2, 4, 6, 8, 5],
    'z': [1, 3, 2, 7, 4]
}

base = pd.DataFrame(dados)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,figsize=(10,5))

ax[0].bar(base.x,base.y, label="y",width=1)
ax[1].plot(base.x,base.y, label="z")                        

                       
plt.legend()
plt.show()
```

```python
#colocando uma batta uma do lado da outra 
fig, ax = plt.subplots()

w = 0.4
ax.bar(base.x+ w/2,base.y, label="y",width=w)
ax.bar(base.x-w/2 ,base.z, label="z",width=w)                        

                       
plt.legend()

plt.show()
```

```python
#colocando texto do grafico
fig, ax = plt.subplots()

w = 0.4
ax.bar(base.x+ w/2,base.y, label="y",width=w)
ax.bar(base.x-w/2 ,base.z, label="z",width=w)                        

#primeiro vc coloca o texto que vc quer depois acordenada (x,y respectivaamente )
for i in range(0,6):
    ax.annotate(base.z[i],(base.x[i]-w/2,base.z[i]))
    ax.annotate(base.z[i],(base.x[i]+w/2,base.y[i]))

                       
plt.legend()

plt.show()
```
