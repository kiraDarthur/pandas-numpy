```python
import numpy as  np
x = np.linspace(0,5 ,11)
```

```python
x
```

```python
y= x **2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
fig =  plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.8,0.8]) #esquerda,inferior,largura,altura
axes2 = fig.add_axes([0.2 ,0.5 ,0.4,0.3]) #esquerda,inferior,largura,altura

axes.plot (x,y , "blue")
axes.set_xlabel('label x')
axes.set_ylabel('label y')
axes.set_title('Titulo')

axes2.plot (x,y , "red")
axes2.set_xlabel('label x2')
axes2.set_ylabel('label y2')
axes2.set_title('Titulo')
```

# SUBPLOTS

```python
fig , ax =plt.subplots()

ax.plot(x,y, "r--")
ax.set_title('Titulo do nosso subplot')
```

```python
fig , ax = plt.subplots(nrows= 1, ncols=2)
for axis in ax:
    axis.plot(x,y,'r')
```
